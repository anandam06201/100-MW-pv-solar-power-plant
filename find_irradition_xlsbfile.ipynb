```python
df1=df[['Unnamed: 16']].loc[2:1141,:]
```

```python
df1.columns = ['irradition_vales']
import numpy as np

df1['irradition_vales'] = df1['irradition_vales'].apply(lambda x: np.nan if x == 'No Data' else float(x))
```

```python
def irradition(t1,m1,t2,m2):
  time1=t1+(m1/60)
  time2=t2+(m2/60)
  sum=df1['irradition_vales'].loc[2+(time1*60):(time2*60)].sum()
  return sum/60000
```

**Time format example is (6hours,10min,18hours,15min)   for finding break down irradtion**

```python
irradition(6,15,15,0)
```
